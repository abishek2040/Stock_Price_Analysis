```python
import yfinance as yf
import plotly.express as px
import pandas as pd
```

```python
tickers = ["AAPL", "MSFT", "AMZN", "NWG.L", "BARC.L", "HSBA.L", "BP.L", "GSK.L", "VOD.L"]
data = yf.download(tickers, start="2018-01-01", end="2024-01-01")
data.head()
```

```python
# Inspecting the structure of the data
print(data.shape)

# Checking any missing values
print(data.isnull().sum().to_frame().T)
data = data.dropna()

# Printing the shape after dropping missing values
print(data.shape)
```

```python
print(data.describe())

# Getting the highest priced stock among all the stocks in the dataset; 
print("\nHighest closing price among all stocks:")
print(data["Close"].max().max())
```

```python
# Plotting the prices of the stocks over time
fig = px.line(data["Close"], title="Stock Closing Prices Over Time")
fig.show()
fig.show(renderer="png")
```
